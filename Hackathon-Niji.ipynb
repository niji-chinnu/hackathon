### Splitting the data 

```python
Y= df.y
X=df.drop("y",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.30,random_state=24)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

### Scalling the data 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,Y_train)
```

```python
prediction_train=logmodel.predict(X_train)
prediction_test=logmodel.predict(X_test)
```

```python
df_pred = pd.DataFrame(logmodel.predict_proba(X_train))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,auc,roc_curve
print(classification_report(Y_train, prediction_train))
print(accuracy_score(Y_train, prediction_train))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,auc,roc_curve
print(classification_report(Y_test, prediction_test))
print(accuracy_score(Y_test, prediction_test))
```

from Logistic Regression we got test and train score 75 % and 74.02

```python
df_cm= confusion_matrix(y_true=Y_train, y_pred=prediction_train)
pd.DataFrame(df_cm)
```

```python
sns.heatmap(df_cm, annot=True, cbar=False, fmt = 'g', annot_kws={'size':25});
```

```python
# ROC and AUC curve
FPR, TPR, _ = roc_curve(Y_train, prediction_train)
FPR1, TPR1, _ = roc_curve(Y_test,prediction_test)
```

```python
auc_train = auc(FPR, TPR)
print('Train AUC: %0.3f' % auc_train)

auc_test = auc(FPR1, TPR1)
print('Test AUC: %0.3f' % auc_test)
```

```python
plt.figure(figsize=(15,7))
plt.plot(FPR, TPR, label = 'Train AUC: %0.3f'%auc_train)
plt.plot(FPR1, TPR1, label = 'Test AUC: %0.3f'%auc_test)

plt.plot([0,1],[0,1], ls = '--', color = 'black', lw = 0.75)

plt.title('ROC Curve')
plt.legend()
plt.show();
```

From above graph we got train and test accuracy of 74%
